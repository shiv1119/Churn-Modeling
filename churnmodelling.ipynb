```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data = pd.DataFrame(data)
data
```

```python
from sklearn.preprocessing import LabelEncoder 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.model_selection import train_test_split 
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
data['Geography'].fillna(data['Geography'].mode()[0],inplace = True)
data['Gender'].fillna(data['Gender'].mode()[0],inplace = True)
data['Age'].fillna(data['Age'].mean(), inplace = True)
```

```python
data.isnull().any()
```

```python
le = LabelEncoder()
data['Geography'] = le.fit_transform(data['Geography'])
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
x = data.iloc[:,3:13].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state=0)
```

```python
print(f"Test shape of x_train = {x_train.shape} and y_train = {y_train.shape}") 
```

```python
print(f"Test shape of x_test = {x_test.shape} and y_test = {y_test.shape}") 
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
lc = LogisticRegression(random_state=0)
lc.fit(x_train, y_train)
```

```python
y_pred = lc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy = ", acc)
```
